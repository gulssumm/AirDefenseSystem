```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.pt')  # Start from YOLOv8 small pretrained weights
model.train(data='/kaggle/working/roboflow/data.yaml', epochs=10, imgsz=320, batch=64, project='/kaggle/working/roboflow', name='yolov8')
```

```python
metrics = model.val(data='/kaggle/working/roboflow/data.yaml')
print(metrics)
```

```python
results = model.predict(source='/kaggle/working/roboflow/test/images', save=True, conf=0.25)
```
